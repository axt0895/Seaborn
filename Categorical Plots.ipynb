```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-whitegrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(6,5))
sns.regplot(x = tips['total_bill'], y = tips['tip'], line_kws={'color': 'blue'}, scatter_kws={'color':'red'})
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.jointplot(x = tips['total_bill'], y = tips['tip'], kind='scatter')
plt.show()
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
plt.figure(figsize=(7,5))
sns.jointplot(y = diamonds['carat'], x = diamonds['price'], kind = 'hex', gridsize = 15)
plt.show()
```
